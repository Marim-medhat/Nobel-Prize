## Dataset Overview

> This dataset includes a record for every individual that was awarded the Nobel Prize since 1901. The Nobel Prize is a set of annual international awards bestowed in several categories by Swedish and Norwegian institutions in recognition of academic, cultural, or scientific advances. Between 1901 and 2020, the Nobel Prizes and the Prize in Economic Sciences were awarded to 949 people . The Nobel Prize is an international award administered by the Nobel Foundation in Stockholm, Sweden, and based on the fortune of Alfred Nobel, Swedish inventor and entrepreneur.

>there are 923 raw and 14 columns about Nobel prize winner information. firstname surname borncountrycode diedcountrycode gender year category share nameofuniversity cityofuniversity countryofuniversity born_month age agegetprize


The datase: https://www.kaggle.com/bahramjannesarr/nobel-prize-from-1901-till-2020

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe

df = pd.read_csv('nobel_final.csv')
df['decade'] = (df.year / 10).astype(int) * 10
df.head(3)
```

## Distribution of the ages

>  the distribution of the ages seem to be normally distributed, it shows that most people who win a Nobel Prize are between their 60s and 70s

```python
plt.hist(df['age_get_prize']);
plt.xlim(0, 100);
plt.xlabel('age when they get Nobel Prize');
plt.ylabel('awarded people Count')
plt.title("ages distribution", fontsize=22);
```

##  Nobel Prize gender

> It appears that about six percent of Nobel Prize winners are women. I expected that the percentage of women would be less than men, but I did not expect that the difference would be this big.

```python

plt.pie(df['gender'].value_counts(), labels=df['gender'].value_counts().index,counterclock=False,autopct='%.2f')
plt.title('awarded people BY GENDER');
```

## Awarded people by category
>The Nobel Prize is given more to people in the field of medicine then physics.



```python
base_color = sb.color_palette()[0]
order= df['category'].value_counts().index
sb.countplot(data=df,y='category',order=order, color=base_color)
plt.title('awarded people BY category',size=20);
```

## Distribution of Nobel Prize by country
>The United States has the Most of the Nobel Prizes.
USA, Great Britain, Deutschland(Germany) had the most wins.


```python
plt.figure(figsize=[20,30])

order= df['born_country_code'].value_counts().index

sb.countplot(data=df,y='born_country_code',order=order, palette = 'CMRmap')
plt.title('awarded people BY COUNTRY',size=20);
```

## The number of men and women over the decades
>over the years the proportion of women increases.

```python
plt.figure(figsize=[10,5])

sb.countplot(data=df,x='decade',hue='gender',palette = 'Greens')
plt.title('Number of men and women  by decade',size=20);
```

## Number of men and women  by category
>Here we can see that women are interested in (peace&literature) then medicine
 but men get a lot of nobel prize at physics and medicine,  but men get a lot of nobel prize at physics and medicine .


```python
order= df['category'].value_counts().index

sb.countplot(data=df,y='category',order=order,hue='gender',palette = 'Greens')
plt.title('Number of men and women  by category',size=20);
```

## Number of women  by Country
>The United States has the Most of the woemn how win Nobel Prizes


```python
order= df['born_country_code'].value_counts().index

plt.figure(figsize=[20,40])
sb.countplot(data=df,y='born_country_code',order=order,hue='gender')
plt.title('Number of men and women  by Country',size=20);
```

# category by Country

>It shows us that most of America's prizes are in medicine and the fewest are in physics.

```python

plt.figure(figsize=(20, 30))
order= df['born_country_code'].value_counts().index

plt.figure(figsize=[20,40])
sb.countplot(data=df,y='born_country_code',order=order,hue='category')
plt.title('number of category by Country',size=20);
```

##  Every winner age seperated by gender and prize category
>This shows that a very small percentage of women won the Nobel Prize in (physics,chemistry, economics). Many young ladies have won the Nobel Prize in peace before the age of thirty-five.



```python

plt.figure(figsize=(20, 5))

sb.swarmplot(x='category', y='age_get_prize',hue = 'gender', dodge=True , data=df, palette = 'CMRmap')
plt.ylabel('Age when get prize')
plt.xlabel('Gender')
plt.title('Every winner age seperated by gender and prize category')

plt.show()
```

##  Every winner gender seperated by decades and prize category
>This shows that only from the sixties did the Nobel Prize awarded to people in the field of economics. There are only two women who have been awarded the Nobel Prize in Economics and have received it in the new millennium.

```python
plt.figure(figsize=(15, 5))
sb.swarmplot(x='gender', y='decade',hue = 'category', dodge=True , data=df)
plt.ylabel('decades')
plt.xlabel('Gender')
plt.title('Every winner gender seperated by decades and prize category')
plt.show()
```
